# Import needed libraries

Firstly, we will be importing these 4 basic libraries
- datetime : dealing with our time series stock data
- matpltlib : dealing with data visualisation
- pasdas : dealing with data mannagment e.g DataFrame , Series
- numpy : dealing with multi-dimensional arrays , mathematical functions and lenear algebra calcualtion

```python
import datetime as dt
import pandas as pd
import pandas_datareader.data as web

import matplotlib.pyplot as plt
%matplotlib notebook
```

---
Read from yahoo. 
The first column will be added to the dataset as an index.

```python
start = dt.datetime(2000,1,1)
end = dt.datetime(2005,12,31)

df_web = web.DataReader('TMB.BK', 'yahoo', start, end)
df_web.head()
```

---
Read from .csv file.
Need to set the index column to date.

```python
df = pd.read_csv('TMB.csv', parse_dates = True, index_col = 0)
df.head()
```

# Create figure

```python
fig, ax = plt.subplots()
ax.plot(df['Open'], color = 'green', label = 'Open price')
ax.plot(df['Close'], color = 'red', label = 'Close price')

plt.xlabel('Trading day')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
```

# Scatter plot

```python
fig, ax = plt.subplots()

ax.scatter(df.index[:20], df['Open'][:20], color = 'green', label = 'Open price')
ax.scatter(df.index[:20], df['Close'][:20], color = 'red', label = 'Close price')

plt.xlabel('Trading day')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
```

# Candlestick plot

```python
from matplotlib.finance import quotes_historical_yahoo_ohlc, candlestick_ohlc
```

```python
start = dt.datetime(2000,1,1)
end = dt.datetime(2005,12,31)

quotes = quotes_historical_yahoo_ohlc('TMB.BK', start, end)

if len(quotes) == 0:
    raise SystemExit
    
fig, ax = plt.subplots()
candlestick_ohlc(ax, quotes, width = 0.6, colorup ='g', colordown = 'r')

ax.xaxis_date()

plt.show()
```
