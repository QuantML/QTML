# Defining time period

```python
start = dt.datetime(2000, 1, 1)
end = dt.datetime(2005, 12, 31)
```

# Download data into python dataframe

```python
df = web.DataReader('TMB.BK', 'yahoo', start, end)
```

# Visualing data
- Date will be an index of dataset by default. 
- head() and tail allow you to visualise only top and bottom of you dataFrame

```python
df.tail(10)
```
